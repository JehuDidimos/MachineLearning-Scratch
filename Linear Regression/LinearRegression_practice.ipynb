```python
from LinearRegression_class import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

dataset = pd.read_csv("TSLA.csv")
X = dataset[["High", "Open", "Low", "Volume"]]
y = dataset["Adj Close"]
```

```python
# sc_X = StandardScaler()
# X = sc_X.fit_transform(X)
```

```python
def mse(true, prediction):
    return np.mean((true - prediction) ** 2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
linear_model = LinearRegression(learning_rate=0.65, n_iterations=10)
linear_model.fit(X_train, y_train)
prediction = linear_model.predict(X_test)
```

```python
print(prediction)
print(mse(y_test, prediction))
```

```python
cmap = plt.get_cmap('viridis')
fig = plt.figure(figsize=(8,6))
plt.plot(y_test, color='black')
plt.plot(prediction, color='purple')
plt.show()
```
